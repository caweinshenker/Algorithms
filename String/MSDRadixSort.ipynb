```python
import random
import time

class MSDRadixSort(object):
    
    @staticmethod
    def sort(arr, R):
        return MSDRadixSort.__sort(arr, R, 0, 0, len(arr))
    
    @staticmethod
    def __sort(arr, R, d, lo, hi):
        
        n = hi - lo
        if n <= 1:
            return
        #print(arr[lo:hi])
        
        aux = [None for i in range(n)]        
        count = [0 for i in range(R+1)]
        
        #Compute frequencies
        for i in range(lo, hi):
            count[ord(MSDRadixSort.__getChar(arr[i], d)) + 1] += 1
            
        #Cumulates
        for r in range(R):
            count[r + 1] += count[r]
        countTmp = count[:]
            
        #Map inputs to outputs
        for i in range(lo, hi):
            s = arr[i]
            c = ord(MSDRadixSort.__getChar(s, d))
            aux[count[c]] = s
            count[c] += 1
        arr[lo:hi] = aux[:]
        
        #Sort subfiles
        for r in range(R):
            MSDRadixSort.__sort(arr, R, d+1, lo + count[r], lo +count[r+1])
        return arr
            
    @staticmethod
    def __getChar(s, d):
        return s[d] if len(s) > d else chr(0)
            
if __name__ == "__main__":
    trials = 100
    for trial in range(trials):
        length = random.randint(1, 100)
        arr = []
        R = 256
        for i in range(length):
            strLength = random.randint(1, 10)
            arr.append("".join([chr(random.randint(65, 90)) for c in range(strLength)]))
        sortedArr = sorted(arr)
        radixSorted = MSDRadixSort.sort(arr, 256)
        for i in range(len(arr)):
            assert sortedArr[i] == radixSorted[i]      

        
```
