```python
from UnionFind import UnionFind
import random
import sys

if __name__ == "__main__":
    nodes = random.randint(1, 10)
    unions = random.randint(0, int(nodes/2))
    print(nodes, unions)
    uf = UnionFind(nodes)
    for i in range(unions):
        u, v = random.randint(0, nodes - 1), random.randint(0, nodes - 1)
        uf.union(u, v)
        assert(uf.connected(u, v))
        assert(uf.connected(uf.find(u), uf.find(v)))
        
        
    uf.display()
    
```
