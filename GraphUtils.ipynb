```python
import sys
import matplotlib.pyplot as plt
import networkx as nx
import random
import pprint


class Edge(object):
    
    def __init__(self, one, other):
        self.one = one
        self.other = other
    
    def __str__(self):
        return "{} <--> {}".format(self.one)
    
    def either(self):
        return self.one
    
    
    def other(self, node):
        if node == self.one:
            return self.other
        elif node == self.other:
            return self.one
        else:
            raise ValueError("Invalid endpoint")

class DirectedEdge(object):
    
    def __init__(self, f, t):
        self.f = f
        self.to = to
        
    def f(self):
        return self.f
    
    def t(self):
        return self.t
        
class Graph(object):
    
    def __init__(self, data=[]):
        self.data = data
        self.adj = [[] for i in range(len(data))]
        self.V = len(data)
        self.E = 0
        
    def __str__(self):
        return """
        Nodes: {} 
        
        Edges:
        {}
        """.format(pprint.saferepr(self.data), pprint.saferepr(self.adj))
    
            
    @staticmethod
    def generateRandomGraph(nodes=None, edges=None):
        g = Graph()
        for i in range(nodes):
            g.addNode(i)
        for i in range(edges):
            t, f = random.randint(0, nodes-1), random.randint(0, nodes-1)
            if t == f or f in g.adjacent(t):
                continue
            g.addEdge(t, f)
        return g
    

    def V(self):
        return self.V
    
    def E(self):
        return self.E
    
    def addNode(self, data):
        self.data.append(data)
        self.adj.append([])
        self.V += 1
        
    def addEdge(self, v, w):
        if v >= self.V or w >= self.V:
            raise ValueError("Invalid edge")
        self.adj[v].append(w)
        self.adj[w].append(v)
        self.E += 1
    
    def adjacent(self, v):
        if v >= self.V:
            raise ValueError("Invalid edge")
        return self.adj[v]
    
    def degree(self, v):
        if v >= V:
            raise ValueError("Invalid edge")
        return len(self.adj[v])
    
        
    def displayGraph(self):
        G = nx.Graph()
        for d in self.data:
            G.add_node(d)
        for i, n in enumerate(self.adj):
            for e in n:
                G.add_edge(i, e)
        nx.draw(G)
        nx.draw_networkx_labels(G, pos=nx.spring_layout(G))
        plt.show()
    
        

if __name__ == "__main__":
    nodes = random.randint(0, 15)
    edges = random.randint(0, nodes * int((nodes - 1) / 2))
    g = Graph.generateRandomGraph(nodes, edges)
    print(nodes, edges)
    print(g)
    g.displayGraph()
        
    
    
    
```
